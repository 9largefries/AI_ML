```python
import cv2
```

```python
im = cv2.imread('souce/1.jpg')

final_wide = 32*13
#коэф уменьшени сторон
r = float(final_wide) / im.shape[1]
dim = (final_wide, int(im.shape[0] * r))

resized = cv2.resize(im, dim, interpolation = cv2.INTER_AREA)

im = resized

wb = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)

row = 0
nfile = 'n/'
efile = 'e/'
ufile = 'u/'
ofile = 'o/'
x0, y0, x1, y1 = 0, 0, 32, 32
size = 32
name = 1
while row < (wb.shape[0] // size):
    column = 0
    x0 = 0
    x1 = x0 + size
    while column < (wb.shape[1] // size):
        if name <= 100:
            file = efile+'{}.jpg'.format(name)
        elif name > 100 and name <= 200:
            file = ufile+'{}.jpg'.format(name)
        else:
            file = ofile+'{}.jpg'.format(name)
        chop = wb[y0:y1, x0:x1]
        cv2.imwrite(file, chop)
        column += 1
        x0 += size
        x1 = x0 + size
        name += 1
    row += 1
    y0 += size
    y1 += size
```
